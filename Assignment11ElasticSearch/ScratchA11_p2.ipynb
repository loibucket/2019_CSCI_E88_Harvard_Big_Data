```python
import pandas as pd
import numpy as np
import random
```

```python
id_list = list(range(0,21))

df = pd.DataFrame(id_list)
df[0] = df[0].astype(str).apply(lambda x: 'id'+x)

#generate incremental time
time_list = [pd.Timestamp('2019-11-22T00Z')]
for i in range(1,len(df)):
    time_list.append( time_list[-1] + pd.Timedelta(seconds=random.randint(3,410)) )
df[1] = time_list

#generate url
new_url_list = ['http://google.com/cow','http://bing.com/cat','http://espn.com/ball','http://espn.com/ball','http://amazon.com/fire']
df[2] = np.random.choice(new_url_list, size=len(df))

#generate user
df[3] = np.random.choice(['bob','carly','moo','bigbig','buck',], size=len(df))

#generate country
df[4] = np.random.choice(['ER','SJ','MA','GD','ZW'], size=len(df))

#generate browser
new_browser_list = ['Chrome','Firefox','Edge','IE']
df[5]  = np.random.choice(new_browser_list, size=len(df))

#generate OS
df[6] = np.random.choice(['Mac','Linux','iPhone','Windows'], size=len(df))

#generate Response
df[7] = np.random.choice([501,307,510,208], size=len(df))

#generate TTFB
df[8] = np.random.uniform(size=len(df))
df[8] = df[8].round(3)

#df
```
