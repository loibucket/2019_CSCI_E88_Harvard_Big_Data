```python
import findspark
findspark.init()

import pyspark
from pyspark import SparkContext
```

```python
sc=SparkContext(master="local[4]")
```

```python
logs = sc.textFile("input_files/*.csv")
```

```python
logs.count()
```

```python
logs.distinct().count()
```

```python
logs.take(1)
```

```python
#read through lines, keep only the dateHour and url
def getDateHourUrl(line):
    columns = line.split(',')
    dateHour = columns[1].split(':')[0]
    url = columns[2]
    return ( dateHour, url )
```

```python
dateHourUrlCollection = logs.map(getDateHourUrl)
```

```python
dateHourUrlCollection.take(2)
```

```python
dateHourUrlCountList = dateHourUrlCollection.map(lambda x: (x[0], x[1]))\
    .groupByKey()\
    .mapValues(lambda vals: len(set(vals)))\
    .sortByKey()\
    .collect()
```

```python
dateHourUrlCountList[:5]
```

```python
f = open("spark_q1_notebook.csv", "w")
for item in sorted( dateHourUrlCountList ):
    f.write(   "%s,%s\n" % (item[0], str(item[1]))   )
```
